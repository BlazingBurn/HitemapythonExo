### 3) En créant un nuage de points, regardez comment se comporte la colonne LotArea par rapport au SalesPrices

```python
# panda
dataHousePricing.plot.scatter("LotArea", "SalePrice", marker='o', c='blue')
```

```python
# Apply the default theme
sns.set_theme()

# Create a visualization
sns.relplot(data=dataHousePricing,y="SalePrice",x="LotArea", height=6).set(title='Nuage de points - LotArea par rapport au SalePrice')
```

### 4) Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$

```python
# Create a visualization
sns.relplot(data=dataHousePricing[(dataHousePricing["SalePrice"] < 500000) & (dataHousePricing["LotArea"] < 20000)],y="SalePrice",x="LotArea",height=6,s=40).set(title="Nuage de point - inférieur 20000pieds et prix 500000€")
```

### 5) En créant un nuage de points, regardez la relation entre le LotFrontage et le LotArea

```python
# Create a visualization
sns.relplot(data=dataHousePricing,x="LotFrontage",y="LotArea", height=6).set(title="Nuage de point - relation LotFrontage et LotArea")
```

### 6) De la même manière, affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotFrontage inférieur à 200 pieds carrés et un LotArea inférieur à 100000 pieds carré

```python
# Create a visualization
plt.figure(figsize=(20,5))

sns.relplot(data=dataHousePricing[(dataHousePricing.LotFrontage < 200) & (dataHousePricing.LotArea < 100000)],x="LotFrontage",y="LotArea", height=6).set(title="Nuage de point - inférieur LotArea 100000 pieds² et LotFrontage 200 pieds²")
```

### 1) Importez les librairies qu’il vous faut

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataSalePrediction = pd.read_csv("sales_predictions.csv")

dataSalePrediction
```

#### **Exemple de problématique pour ce dataFrame : Le temps augmente t-il le prix d'un objet ?**

### 2) En utilisant relplot(), construisez un graphique qui va vous permettre de voir l’évolution des prix par rapport au temps. Que pouvez vous voir ?

```python
# Create a visualization
sns.relplot(data=dataSalePrediction,x="item_price", y="date", height=6).set(title="Nuage de point - evolution du prix par rapport au temps")
```

### 3) Corrigeons le problème de visualisation, en utilisant la fonction .sample() de Pandas, prenez un échantillons de 50 éléments dans votre dataset

```python
# Create a visualization
sns.relplot(data=dataSalePrediction.sample(n=50, random_state=1),x="item_price", y="date", height=7).set(title="Nuage de point - evolution du prix par rapport au temps (50 éléments)")
```

### 4) Retentez de faire votre visualisation, créez une figure de taille (20,6). Que voyez vous ?

```python
# Create a visualization
sns.relplot(data=dataSalePrediction.sample(n=50, random_state=1),x="item_price", y="date", height=6, sizes=(20,6)).set(title="Nuage de point - evolution du prix par rapport au temps (50 éléments)")
```

### 5) En utilisant la fonction pd.to_datetime(), convertissez votre colonne date en datetime

```python
datime = pd.to_datetime(dataSalePrediction['date'])
print(datime)
sns.relplot(data=dataSalePrediction.sample(n=50, random_state=1),y="item_price", x=pd.to_datetime(dataSalePrediction["date"]), height=6, sizes=(20,6)).set(title="Nuage de point - evolution du prix par rapport au temps (50 éléments) | date to datetime")
```

### 6) Retentez une dernière fois votre visualisation.

```python
# Create a visualization
sns.lineplot(data=dataSalePrediction.sample(n=50, random_state=1),y="item_price", x=pd.to_datetime(dataSalePrediction["date"])).set(title="Nuage de point - evolution du prix par rapport au temps (50 éléments) | date to datetime")
```
